```python
import pandas

df = pandas.read_csv('data/wordle-counts.csv', parse_dates=['start', 'end'])
df
```

```python
df = df.sort_values('start')
df
```

```python
import plotly.express as xp

xp.line(df, x='start', y='day_count')
```

```python
from datetime import datetime, timezone

recent = df[(df['start'] > datetime(2021, 10, 1, tzinfo=timezone.utc)) & (df['start'] < datetime(2022, 1, 27, tzinfo=timezone.utc))]
xp.line(recent, x='start', y='day_count', title='Worldle Tweets per Day', labels={'day_count': 'Tweets per Day', 'start': 'Day'})
```
