# COVID Vaccinations by Race (Maryland)

```python
import pandas

pops = pandas.read_excel('https://planning.maryland.gov/MSDC/Documents/pop_estimate/ARS/table1a.xlsx')
pops
```

```python
pops = pops[[f'Unnamed: {n}' for n in range(0, 12)]]
pops = pops.drop(['Unnamed: 2'], axis='columns')
pops
```

```python
pops = pops.drop([0, 1, 2, 3, 4])
pops
```

```python
pops = pops.drop(range(11, 34))
pops
```

```python
pops.columns = ["race", *[str(n) for n in range(2010, 2020)]]
pops
```

```python
pops['race'] = pops['race'].apply(lambda s: s.strip('.').upper())
pops
```

```python
pops.loc[6]['race'] = 'BLACK OR AFRICAN AMERICAN'
pops.loc[7]['race'] = 'AMERICAN INDIAN OR ALASKAN NATIVE' 
pops.loc[9]['race'] = 'NATIVE HAWAIIAN OR OTHER PACIFIC ISLANDER'
pops
```

```python
pops = pops.melt(['race'], var_name='year', value_name='population')
pops['population'] = pops['population'].apply(int)
pops
```

```python
pops['year'] = pops['year'].apply(int)
pops
```

```python
pops.to_csv('data/md-population-race.csv')
```

```python
vaccs = pandas.read_csv('https://opendata.arcgis.com/datasets/7ca637ad378647e084e01566591d3d13_0.csv', parse_dates=['VACCINATION_DATE'])
vaccs
```

Remove data for before there were vaccines.

```python
import datetime

vaccs = vaccs[vaccs["VACCINATION_DATE"] >= datetime.datetime(2020, 12, 11, tzinfo=datetime.timezone.utc)]
vaccs
```

```python
vaccs = vaccs.drop('OBJECTID', axis='columns')
vaccs
```

```python
vaccs.columns = ["date", "race", "first_dose_daily", "first_dose_cumulative", "second_dose_daily", "second_dose_cumulative", "single_dose_daily", "single_dose_cumulative"]
vaccs
```

```python
vaccs = vaccs[~ vaccs['race'].isin(['UNKNOWN', 'OTHER', "DATA NOT AVAILABLE"])]
vaccs
```

```python
vaccs.race.value_counts()
```

```python
#vaccs = vaccs[["date", "race", "first_dose_daily", "second_dose_daily", "single_dose_daily"]]
#vaccs
```

```python
vaccs = vaccs.fillna(0)
vaccs
```

```python
vaccs['date'] = vaccs['date'].apply(lambda d: d.date)
vaccs
```

```python
vaccs.date.min()
```

```python
vaccs.to_csv('data/md-vaccinations-race.csv')
```

## Create SQLite

Create a sqlite db for students to work with for the exercise.

```python
import sqlite3

db = sqlite3.connect('data/md-vaccinations.sqlite3')

vaccs.to_sql(name="vaccinations", index_label="id", con=db)
pops.to_sql(name="populations", index_label="id", con=db)
```

```python
pops_2019 = pops[pops['year'] == 2019]
pops_2019
```

```python
pops_2019 = pops_2019.drop('year', axis='columns')
pops_2019
```

```python
vaccs_pops = pandas.merge(vaccs, pops_2019, on='race')
vaccs_pops
```

```python
vaccs_pops['date'].min()
```

```python
vaccs_pops['vaccinated'] = vaccs_pops['second_dose_cumulative'].fillna(0) + vaccs_pops['single_dose_cumulative'].fillna(0)
vaccs_pops
```

```python
vaccs_pops['vaccinated_percent'] = vaccs_pops['vaccinated'] / vaccs_pops['population']
vaccs_pops
```

```python
vaccs_pops = pandas.pivot_table(vaccs_pops, columns='race', values='vaccinated_percent', index='date')
vaccs_pops
```

```python
vaccs_pops = vaccs_pops.fillna(0)
vaccs_pops
```

```python
from plotly import express as px

layout = {}
fig = px.line(vaccs_pops, labels={'date': 'Date', 'race': 'Race', 'value': 'Percent of Population (Census)'},
    title='COVID Vaccinations by Race (Maryland)')
fig
```

```python
fig.write_html('data/md-vaccinations-race.html')
```
