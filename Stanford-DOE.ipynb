# Stanford and DoE

> Tom: What can RIALTO tell us about what level of DOE funding has created Stanford / SLAC research outputs?

```python
import pandas

df = pandas.read_csv('../sul-dlss/rialto-airflow/data/snapshots/20241114173510/dimensions-pubs.csv')
df
```

Only include publications that have funding infrormation.

```python
df = df[df.funders.notna()]
df
```

```python
266787 / 537673
```

Only half of our Dimensions publications have funding information.

Parse the funders information, so we can get the DoE ones. There can be more than one funder per publication.

```python
eval(df['funders'][0])
```

Lets get the acronyms.

```python
df['funder_acronyms'] = df.funders.apply(lambda funders: [f.get('acronym') for f in eval(funders)])
df
```

```python
doe = df[df.funder_acronyms.apply(lambda f: 'DOE' in f)]
doe
```

```python
8814 / 266787
```

Only 3.3% of funders includes the Department of Energy -- assuming that DOE is consistently used.

```python
doe_counts = doe.value_counts('year').reset_index().sort_values('year', ascending=False)
doe_counts
```

```python
doe_counts['year'] = doe_counts['year'].astype('int')
doe_counts
```

```python
from plotly import express as xp

fig = xp.line(
    doe_counts,
    x='year',
    y='count',
    line_shape='spline',
    title='Department of Energy Funded Publications at Stanford',
    labels={'count': 'Publications', 'year': 'Year'},
    width=1000
)
fig.show('png')
```
