## RoR IDs

Maybe it's a small number of DoE publications because some funders within the DoE are listed as the funder, rather than the DoE itself?

For example [The Office of Environmental Management](https://ror.org/00km40770) is within the DoE.

Is it possible to get a list of all the funders withink the DoE and add them? ALso perhaps we should be using the ID to match?

```python
df['funder_ids'] = df.funders.apply(lambda funders: [f.get('id') for f in eval(funders)])
df[['doi', 'funder_ids']]
```

Now we need a list of funder IDs that are part of the DoE. Dimensions uses GRID identifiers so lets create a function that uses the ROR API and returns the GRID IDs for a given ROR ID and all its children that have GRID IDs.

```python
import requests
import time

def grid_ids(ror_id):
    time.sleep(.25)    
    ror_id = ror_id.split('/')[-1] # remove URL portion if present
    
    org = requests.get(f'https://api.ror.org/v2/organizations/{ror_id}').json()

    # if the org has a gridid return it
    grid_id = get_grid_id(org)
    if grid_id:
        yield grid_id

    # look for gridids in child organizations
    for child_org in filter(lambda o: o['type'] == 'child', org['relationships']):
        yield from grid_ids(child_org['id'])

def get_grid_id(org):
    for id in org['external_ids']:
        if id['type'] == 'grid':
            return id['preferred']

doe_grid_ids = list(grid_ids('https://ror.org/01bj3aw27'))
doe_grid_ids
```

Now we can look for publications with a funder that includes one of these.

```python
doe = df[df.funder_ids.apply(lambda ids: len(set(ids) & set(doe_grid_ids)) > 0)]
doe
```

```python
doe_counts = doe.value_counts('year').reset_index().sort_values('year', ascending=False)
doe_counts['year'] = doe_counts['year'].astype('int')
doe_counts = doe_counts[doe_counts.year != 2025]

fig = xp.line(
    doe_counts,
    x='year',
    y='count',
    line_shape='spline',
    title='Department of Energy Funded Publications at Stanford',
    labels={'count': 'Publications', 'year': 'Year'},
    width=1000,
    height=600
)
fig.show('png')
```
