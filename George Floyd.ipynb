# George Floyd

I've been collecting tweets with the hashtag #GeorgeFloyd following the murder of George Floyd. I counted the number of tweets per hour and put them into a CSV.

```python
import pandas

df = pandas.read_csv('data/georgefloyd.csv', parse_dates=['time'])
df.head()
```

```python
import altair

altair.renderers.enable('notebook')

chart = altair.Chart(df, width=800, title="#GeorgeFloyd (Tweets per Hour)")
chart = chart.configure_axisX(labelAngle=45)

chart = chart.mark_line().encode(
    altair.X('time', title="Time"),
    altair.Y('tweets', title="Tweets per Hour"),
)

chart = chart.configure_mark(color='#095')

chart
```

```python
chart = altair.Chart(df, width=600, title="#GeorgeFloyd")

chart = chart.mark_bar(width=25).encode(
    altair.X('monthdate(time):T', title="Time"),
    altair.Y('tweets', title="Tweets per Day"),
)

chart
```
